```python
from datetime import date, timedelta
import polars as pl
from polars_bloomberg import BQuery

universe = [
    "USDCNH Curncy", "KWN+1M Curncy", "USDSGD Curncy", "IDN+1M Curncy",
    "INH+1M Curncy", "IRN+1M Curncy", "MRN+1M Curncy", "PPN+1M Curncy",
    "USDTHB Curncy", "USDJPY Curncy", "ESZ5 Index", "NQZ5 Index",
    "TY1 Comdty", "VIX Index", "XAUUSD Curncy", "CL1 Comdty", "DXY Index",
]

# optional: de-dup if you accidentally include the same ticker twice
universe = list(dict.fromkeys(universe))

end_date = date.today()
start_date = end_date - timedelta(days=365 * 3)

with BQuery() as bq:
    df = bq.bdh(
        universe,
        ["PX_LAST"],
        start_date=start_date,
        end_date=end_date,
    )

# df is a Polars DF: [security, date, PX_LAST]
pivot = (
    df.pivot(values="PX_LAST", index="date", columns="security")
      .sort("date")
)

print(pivot.head())
```
